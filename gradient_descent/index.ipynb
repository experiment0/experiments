```python
import numpy as np
from sympy import *
%matplotlib inline

from classes.GradientDescent import GradientDescent
```

Зададим начальные условия.

```python
# Функция от двух переменных
x, y = symbols('x y', real=True)

# function = x**2 + y**2
function = (1 - x)**2 + 100*(y - x**2)**2

# Запускаем работу градиентного спуска
gradient_descent = GradientDescent(
    function, 
    x, 
    y,
    point_start = np.array([-2, -2]),
    min_acceptable_gradient = 0.1,
    # should_display_logs=true,
)
gradient_descent.run()
```
